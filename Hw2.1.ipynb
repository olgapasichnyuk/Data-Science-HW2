```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
table = pd.read_html(io='https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8',
            match='Коефіцієнт народжуваності')

table
```

```python
table = table[0]
table
```

```python
fig, axs = plt.subplots()

axs.plot(table['Регіон'], table['1990'], color="red", label='1990')
axs.plot(table['Регіон'], table['2000'], color="green", label='2000')
axs.plot(table['Регіон'], table['2012'], color="blue", label='2012')

plt.xticks(rotation=90)
plt.xlabel("Регіон", fontsize=11, color="gray")
plt.ylabel('Коефіцієнт', fontsize=11, color="gray")
plt.title('Коефіцієнт народжуваності в регіонах України в 2012 році', fontsize=14, loc='center')
plt.legend()
plt.show()
```

```python
#Замініть у таблиці значення "—" на значення NaN

table.replace(to_replace='—', value=np.nan, inplace=True)
table
```

```python
sns.set_style("whitegrid")
for i in table.columns[1:]:
    sns.scatterplot(data=table, x=i, y="Регіон")

```

```python
#Визначте типи всіх стовпців за допомогою dataframe.dtypes

table.dtypes
```

```python
for col in table.columns:
    table[col] = pd.to_numeric(table[col], errors='ignore')
table.dtypes
```

```python
#Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)

#for col in table.columns[1:]:

spaces = table.isnull()
res = pd.DataFrame()
res['amount_of_spaces'] = np.sum(spaces, axis=0)
res['part_of_spaces_%'] = (np.sum(spaces, axis=0) / spaces.shape[0]) *100
res[1:]
```

```python
#Видаліть з таблиці дані по всій країні, останній рядок таблиці
table = table[:-1]
table
```

```python
#Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)
table = table.fillna({i: table[i].mean() for i in table.columns[1:]})
table
```

```python
#Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні
table[table['2019'] > table['2019'].mean()]['Регіон']
```

```python
labels = table[table['2019'] > table['2019'].mean()]['Регіон']

data = table[table['2019'] > table['2019'].mean()]['2019']
plt.title('Регіони, де рівень народжуваності у 2019 році був вищим за середній по Україні', fontsize=16, loc='center')


plt.pie(
    data,
    labels=labels,
    autopct="%.2f%%"
)

plt.show()
```

```python
#У якому регіоні була найвища народжуваність у 2014 році
table[table['2014'] == table['2014'].max()]['Регіон']
```

```python
# Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році
plt.bar(table["Регіон"],table["2014"],
    color=["b", "r", "y", "g"],
)

plt.xlabel("Регіон", fontsize="small", color="midnightblue")
plt.ylabel('Коефіцієнт народжуваності', fontsize='small', color='midnightblue')
plt.title('Коефіцієнт народжуваності в регіонах України', fontsize=12)
plt.xticks(rotation=90)


plt.show()
```
